# Our first example, now with persistence homology

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import gudhi as gd

from matplotlib.patches import Polygon
from matplotlib.colors import ListedColormap
from matplotlib.collections import PatchCollection
from sklearn.metrics import pairwise_distances
```

First we make the U and K.

```python
U = np.array([[0,1],[.2,0.8],[.3,1.3],[.1,2.1],[.2,.2],[-0.1,1.7],[.1,.6],[.6,.2],[1.1,0.2],[1.6,0.2],[.8,0],[1,0.3],[2,1.1],[2.1,1.6],[1.9,1.7],[1.8,0.3],[1.95,0.9],[1.5,0.1],[2,2.2]])
K = np.array([[3,1],[3.2,0.6],[3.3,1.2],[3.1,1.8],[3.15,1.9],[3.1,2.1],[3.2,.2],[2.9,1.6],[3.1,.6],[3.6,.6],[4.1,1.2],[4.6,1.7],[3.3,1.5],[4.8,1.9],[5.1,1.8],[4,1],[4.5,0.7],[4.3,1.35],[4.8,0.2],[5,0.3],[5.1,0.1],[3.6,1],[4.9,2]])
UK = np.concatenate((U,K))

plt.plot(UK[:,0],UK[:,1],'o')
plt.axis('equal')
plt.show()
```

```python
D = pairwise_distances(UK)
skeleton = gd.RipsComplex(distance_matrix = D, max_edge_length = 5) 
Rips_complex = skeleton.create_simplex_tree(max_dimension = 2)
```

We now display the barcode diagrams.

```python
BarCodes = Rips_complex.persistence()

gd.plot_persistence_barcode(BarCodes,colormap=['darkorange','blue'])
plt.show()
```

We can instead draw the persistence diagram with the following code.

```python
gd.plot_persistence_diagram(BarCodes,colormap=['darkorange','blue'])
plt.show()
```

The persistent degree 0 homology class appears at height $\infty$. Dots near the diagonal line are considered noise.
