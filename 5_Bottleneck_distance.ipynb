## Displaying the optimal matching

We can also ask it to display the optimal matching. We use the `persim` package for this.

```python
import persim
from persim import plot_diagrams 
```

This seems to work better if we don't use any infinite barcodes.

```python
dgm1 = np.array([
    [0, 4],
    [1, 2],
    [3.2, 3.3],
    [4,4.2]
])
dgm2 = np.array([
    [1, 3],
    [2,3],
    [2,5],
    [4.7,5]
])

d, matching = persim.bottleneck(
    dgm1,
    dgm2,
    matching=True
)

persim.bottleneck_matching(dgm1, dgm2, matching, labels=['First dataset', 'Second dataset'])
plt.title("Bottleneck Distance {:.3f}".format(d))
plt.show()
```

Now here is example 3, which has some infinite barcodes. As you can see, the function does not handle those well, and just ignores them.

```python
dgm1 = np.array([
    [2, 3],
    [2, 3],
    [3, 5],
    [4,float('Inf')]
])
dgm2 = np.array([
    [1, 2],
    [1,4],
    [3,float('Inf')],
])

d, matching = persim.bottleneck(
    dgm1,
    dgm2,
    matching=True
)

persim.bottleneck_matching(dgm1, dgm2, matching, labels=['First dataset', 'Second dataset'])
plt.title("Bottleneck Distance {:.3f}".format(d))
plt.show()
```
